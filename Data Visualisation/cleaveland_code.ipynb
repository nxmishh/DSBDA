```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Cleavland.csv")

# Rename columns for better understanding
df.columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'cholesterol', 'fasting_blood_sugar', 
              'rest_ecg', 'max_heart_rate', 'exercise_induced_angina', 'st_depression', 'st_slope', 'num_major_vessels', 'thalassemia', 'target']

# Convert '?' to NaN
df.replace('?', pd.NA, inplace=True)
df = df.dropna()

# Convert object columns to numeric
df = df.apply(pd.to_numeric)

# Define colors for pie chart
colors = ['#ff9999','#66b3ff']

# a. Histograms
df.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()

# b. Pie Chart for sex distribution
sex_counts = df['sex'].value_counts()
plt.pie(sex_counts, labels=['Male', 'Female'], colors=colors, autopct='%1.1f%%')
plt.title('Sex Distribution')
plt.axis('equal')
plt.show()

# c. Box Plot for age distribution based on chest pain type
plt.figure(figsize=(10, 6))
sns.boxplot(x='chest_pain_type', y='age', data=df)
plt.title('Age Distribution based on Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Age')
plt.show()

# d. Scatter Plot for age vs. max heart rate
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['max_heart_rate'], alpha=0.5)
plt.title('Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()

# e. Scatter Plot with Boxplot for age vs. max heart rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='max_heart_rate', hue='sex', alpha=0.5)
sns.boxplot(data=df, x='age', y='max_heart_rate', color='gray', showfliers=False)
plt.title('Age vs. Maximum Heart Rate with Gender Distribution')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.legend(title='Sex', loc='upper right')
plt.show()
```
