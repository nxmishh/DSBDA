```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Cleavland.csv")
df
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["Age"],bins=20,kde=True)
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.stripplot(x="Sex",y="Age",data=df)
plt.title("Age vs Sex Dot Plot")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Sex",y="Age",data=df)
plt.title("Age vs Sex Bar Plot")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x="Sex",y="Age",data=df)
plt.title("Age vs Sex Line Plot")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["Age"],bins=20,kde=True)
sns.scatterplot(x="Age",y="Sex",data=df)
sns.boxplot(x="Age",y="Sex",data=df)
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
counts=df["Sex"].value_counts()
plt.pie(counts,labels=["Male","Female"],colors=["#f5f521","#f8e1a2"],autopct='%1.1f%%')
plt.title("Sex Ratio")
# plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="Age",y="Sex",data=df)
sns.boxplot(x="Age",y="Sex",data=df)
plt.title("Boxplot to Scatterplot")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
